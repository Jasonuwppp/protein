```python
import pandas as pd
import numpy as np
import xgboost as xgb
from sklearn import svm
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from catboost import CatBoostClassifier
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import StackingClassifier
from sklearn.model_selection import cross_val_score
```

```python
# load data
train = pd.read_csv('../data/train_after_encode.csv')
test = pd.read_csv('../data/test_after_encode.csv')
protein_id = pd.read_csv('../data/protein_id.csv')
# define X,y
X_train = train.drop('y', axis=1)
y_train = train['y']
# create validation sets
#X_train, X_val, y_train, y_val = train_test_split(train_X, train_y, test_size = 0.3)
```

```python
X_train = X_train.astype(np.float64)
y_train = y_train.astype(np.float64)
```

```python
# import best models
lr = pd.read_pickle('../best models/LR.pickle')
catboost = pd.read_pickle('../best models/catboost.pickle')
knn = pd.read_pickle('../best models/KNN.pickle')
lin_svm = pd.read_pickle('../best models/SVM_linear.pickle')
rf = pd.read_pickle('../best models/random_forest.pickle')
lr_bag = pd.read_pickle('../best models/LR_bagging.pickle')
rf_bag = pd.read_pickle('../best models/RF_bagging.pickle')
```

```python
# stacking
level0 = list()
#level0.append(('lr', lr))
#level0.append(('catboost', catboost))
#level0.append(('knn', knn))
#level0.append(('svm', lin_svm))
level0.append(('rf_bag', rf_bag))
level0.append(('lr_bag', lr_bag))
# define meta learner model
level1 = lr_bag
# define the stacking ensemble
stacked = StackingClassifier(estimators=level0, final_estimator=level1, cv=10)
```

```python
# cross validation
scores = cross_val_score(stacked, X_train, y_train, cv=10)
scores.mean()
```

```python
# print info
print("%0.2f accuracy with a standard deviation of %0.2f" % (scores.mean(), scores.std()))
```

```python
# stack train
stacked.fit(X_train, y_train)
```

The following is related to submission

```python
# creating the result vector
y_pred_test = stacked.predict(test).astype(int)
```
