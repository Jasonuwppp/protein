```python
import numpy as np
import pandas as pd
from sklearn.metrics import accuracy_score
#from sklearn.model_selection import train_test_split
from sklearn.model_selection import GridSearchCV
from sklearn.naive_bayes import MultinomialNB
```

```python
# load data
train = pd.read_csv('../data/train_after_encode.csv')
test = pd.read_csv('../data/test_after_encode.csv')
protein_id = pd.read_csv('../data/protein_id.csv')
# define X,y
X_train = train.drop('y', axis=1)
y_train = train['y']
# create validation sets
#X_train, X_val, y_train, y_val = train_test_split(train_X, train_y, test_size = 0.3)
```

```python
X_train.shape
```

```python
y_train.shape
```

```python
# cast the type to float64
X_train = X_train.astype(np.float64)
y_train = y_train.astype(np.float64)
```

```python
# parameter grid
params = {
    'alpha': np.arange(0,4.1,0.1)
}
```

```python
# create a naive bayes classifier
clf = MultinomialNB()
# grid search
grid = GridSearchCV(clf,params,cv=10)
# train the classifier
grid.fit(X_train,y_train)
```

```python
# best parameter
grid.best_params_
```

```python
# best predictor
best_nb = grid.best_estimator_
```

```python
# accuracy score
grid.best_score_
```

```python
# save the linear kernel
pd.to_pickle(best_nb, '../best models/naive_bayes.pickle')
```
