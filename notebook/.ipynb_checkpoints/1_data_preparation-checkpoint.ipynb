## 1. Data preparation

```python
# find what columns contain constant values in test.csv, they should be removed because they cannot insert an influnece on y
remove = test.columns[test.nunique()<=1]
test = test.drop(test.columns[remove], axis=1)
test
```

```python
# do the same thing for train.csv
remove = train.columns[train.nunique()<=1]
train = train.drop(train.columns[remove], axis=1)
# also drop the second last column
train = train.drop(2959,axis=1)
train
```

```python
# we use intersection to find what columns exist in both train and test
test_col = test.columns
train_col = train.columns
intersection = test_col.intersection(train_col)
# slice two df
train = train[intersection]
test = test[intersection]
# append y to train
train['y'] = y_train
```
